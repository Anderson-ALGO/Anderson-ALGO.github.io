Otros parÃ¡metros:
    * sheet_name (que hoja quiero)
    * index_col  (cual columna quiero de indice)
    * names/use_cols  (nombres de columnas en caso de no tener / columnas que quiero usar, en  caso de si tener)

    * skipfooter (columnas a ignorar en fin, se pasa argumento como entero)

# CreaciÃ³n de un DataFrame de cero

Muchas veces nos vamos a ver en la necesidad de ir armando nuestros dataframes a veces de cero e ir llenandolos, y a veces partiendo de listas, diccionarios o estructuras similares, empecemos a ver algunos casos

### CreaciÃ³n de un dataFrame a partir de una lista de listas

```python
# Supongamos que partimos de esta lista de datos:
data = [["ALUA",19.15],["BBAR",73.70],["BMA",144.4],["BYMA",234]]
```

Y ahora quiero armar un dataframe en la variable "tabla" con esa estructura que tiene tickers y sus precios
<br>Bien, es tan sencillo como esto:
<br>1. Asignamos a la variable "tabla" el objeto pandas.DataFrame
<br>2. Pasamos como primer argumento del DataFrame los datos (variable "data")
<br>3. Le indicamos en el argumento "columns" los nombres de las columnas

```python
import pandas as pd
data = [["ALUA",19.15],["BBAR",73.70],["BMA",144.4],["BYMA",234]]
tabla = pd.DataFrame(data, columns = ['Ticker', 'Precio']) 
tabla
```

TambiÃ©n podemos crear el dataFrame de forma muy sencilla a partir de un diccionario
<br>En este caso ni siquiera hace falta que le pasemos los nombres de las columnas ya que los toma del mismo diccionario

### CreaciÃ³n de DataFrames a partir de Diccionarios de listas

```python
data = {'Strings':['ALUA', 'BBAR', 'BMA', 'BYMA'], 'Precios':[19.15,73.7 , 144.4, 234]}
tabla = pd.DataFrame(data) 
tabla
```

### CreaciÃ³n a Partir de una lista de diccionarios

Esta estructura de entrada es mas rara, pero tambiÃ©n es posible recibir datos de esta forma algÃºn dÃ­a y por suerte pandas se encarga de transformar todo esto en un data frame de la misma forma sencilla

```python
data = [
        {"ticker": "ALUA", 'Precio': 19.15, "Tipo": "Accion"}, 
        {"ticker": "BBAR", 'Precio': 73.7, "Tipo": "Accion"},
        {"ticker": "BMA", 'Precio': 144.4},
        {"ticker": "COME", 'Precio': 234, "Tipo": "COME"},
       ]

tabla = pd.DataFrame(data) 
tabla.set_index('ticker', inplace=True)
tabla
```

### CreaciÃ³n a partir de otro dataFrame

Partiendo de un DataFrame se puede armar otro con las columnas especÃ­ficas que le pasemos e incluso en la misma lÃ­nea podemos definir cual va a ser el Ã­ndice del dataFrame nuevo

```python
data = pd.read_excel('excels_csvs/AAPL.xlsx', sheet_name='Hoja1')
copia = data[['timestamp', 'open', 'close']].copy().set_index("timestamp")

copia.head()
```

Algo parecido pero "al revÃ©s" serÃ­a crear la copia a partir del original MENOS las columnas que no queremos

```python
data = pd.read_excel('excels_csvs/AAPL.xlsx', sheet_name='Hoja1')
copia = data.drop(['high', 'low', 'adjusted_close'],axis=1)
copia.head()
```

```python
url = 'http://www.rava.com/precios/panel.php?m=LID'
df = pd.read_html(url, thousands='.')[9]
df.columns = list(df.loc[0,:])
df.drop(0,axis=0)
```

Aca va para las opciones y los cedears con el agregado del atributo "header" y pasando a datos numericos los campos

```python
import pandas as pd
url = 'http://www.rava.com/precios/panel.php?m=OPC'
df = pd.read_html(url, thousands='.',header=0)[8]
df = df.replace(',','.',regex=True)
df[df.columns[1:7]] = df[df.columns[1:7]].apply(pd. to_numeric, errors='coerce').round(2)
df[df.columns[8:10]] = df[df.columns[8:10]].apply(pd. to_numeric, errors='coerce')
df
```

```python
import pandas as pd
url = 'http://www.rava.com/precios/panel.php?m=CED'
df = pd.read_html(url, thousands='.',header=0)[8]
df = df.replace(',','.',regex=True)
df[df.columns[1:7]] = df[df.columns[1:7]].apply(pd. to_numeric, errors='coerce').round(2)
df[df.columns[8:10]] = df[df.columns[8:10]].apply(pd. to_numeric, errors='coerce')
df
```
