#### Ejercicio 
Se requiere concatenar dos cotizaciones historicas por medio del adj close de dos tickers cualquiera, el join deber outer y el axis 1 

```python
import yfinance as yf
import pandas as pd

ggal = yf.download('GGAL')['Adj Close']
ypf = yf.download('YPF')['Adj Close']

historicas = pd.concat([ggal,ypf], join ='outer', axis=1)
historicas.columns = ["GGAL", "YPF"]
historicas.dropna(inplace=True)
historicas

```

####Â Ejercicio 

    - Se requiere descargar la informacion de un activo X y obtener la sma de 21, 42 y 200 ponerla en el dataframe
    - calcular las bandas de Bollinger para dicho activo
    - Agregar una columna mas que indique true cuando la sma de 21 sea mayor a la de 42 y a la de 200, false en los otros casos.

```python
import yfinance as yf
import pandas as pd

data = yf.download('AAPL')['Adj Close']

data = pd.DataFrame(data)


data['sma_21'] = data['Adj Close'].rolling(21).mean()
data['sma_42'] = data['Adj Close'].rolling(42).mean()
data['sma_200'] = data['Adj Close'].rolling(200).mean()

data['indicador'] = ((data['sma_21'] > data['sma_42']) & (data['sma_21'] > data['sma_200']))

data['desvio'] = data['Adj Close'].rolling(20).std()

data['bb_sup'] = data['sma_21'] + 2 * data['desvio']
data['bb_inf'] = data['sma_21'] - 2 * data['desvio']

data.dropna()
```

####Â Ejercicio 
Calculando el dolar CCL, descargar el ultimo mes de cotizacion de GGAL y GGAL.BA y calcula el valor del dolar contado con liquidaciÃ³n, utilizar la columna close para calcularlo, concatenar ambos resultados y agregar dicho valor como una columna extra con el valor del calculo.
La cuenta es GGAL.BA / (GGAL * 10)

```python
import yfinance as yf
import pandas as pd

ggal = yf.download('GGAL')['Adj Close']
ggal_ba = yf.download('GGAL.BA')['Adj Close']

data = pd.concat([ggal, ggal_ba], axis=1)

data.dropna(inplace=True)

data.columns = ["GGAL", "GGAL.BA"]

data["CCL"] = data["GGAL.BA"] / (data["GGAL"] / 10)

data
```
