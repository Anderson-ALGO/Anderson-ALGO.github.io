####Â Ejercicio 

    - Se requiere descargar la informacion de un activo X y obtener la sma de 21, 42 y 200 ponerla en el dataframe
    - calcular las bandas de Bollinger para dicho activo
    - Agregar una columna mas que indique true cuando la sma de 21 sea mayor a la de 42 y a la de 200, false en los otros casos.

```python
import yfinance as yf
import pandas as pd

pam = yf.download('PAM', interval='1d' , start='2019-01-01')

pam['sma_21'] = pam['Close'].rolling(21).mean()
pam['sma_42'] = pam['Close'].rolling(42).mean()
pam['sma_50'] = pam['Close'].rolling(50).mean()

pam['sma_20'] = pam['Close'].rolling(20).mean()
pam['desvio'] = pam['Close'].rolling(20).std()
pam['bb_sup'] = pam['sma_20'] + 2 * pam['desvio']
pam['bb_inf'] = pam['sma_20'] - 2 * pam['desvio']

pam['CondiciÃ³n'] = (pam['sma_21'] > pam['sma_42']) & (pam['sma_21'] > pam['sma_50'])

pam.dropna()
```

####Â Ejercicio 
Calculando el dolar CCL, descargar el ultimo mes de cotizacion de GGAL y GGAL.BA y calcula el valor del dolar contado con liquidaciÃ³n, utilizar la columna close para calcularlo, concatenar ambos resultados y agregar dicho valor como una columna extra con el valor del calculo.
La cuenta es GGAL.BA / (GGAL * 10)

```python
import yfinance as yf
import pandas as pd

ggal_byma = yf.download('GGAL.BA')['Close']
ggal_adr = yf.download('GGAL')['Close']

ggales = pd.concat([ggal_byma, ggal_adr], join = 'inner', axis = 1)
ggales.columns = ['GGAL ARGY', 'GGAL ADR']
ggales['CCL'] = (ggal_byma*10)/ggal_adr

ggales.dropna()
```
