####Â Ejercicio 

    - Se requiere descargar la informacion de un activo X y obtener la sma de 21, 42 y 200 ponerla en el dataframe
    - calcular las bandas de Bollinger para dicho activo
    - Agregar una columna mas que indique true cuando la sma de 21 sea mayor a la de 42 y a la de 200, false en los otros casos.

```python
X = yf.download ('X') 

X['sma_21'] = X['Close'].rolling(21).mean()
X['sma_42'] = X['Close'].rolling(42).mean()
X['sma_200'] = X['Close'].rolling(200).mean()

'''
Los valores por defecto que se utilizan para su cÃ¡lculo son de 21 para la media y 2 desviaciones estÃ¡ndar. 
Si se reduce o se incrementa de forma significativa el valor de la media hay ajustar en el mismo sentido el nÃºmero 
de desviaciones estÃ¡ndar. Para valores de la media por encima de 50 (largo plazo) => 2,5 desviaciones; 
valores de la media cercanos a 10 (corto plazo)=> 1,5 desviaciones.
'''
X['desvio'] = X['Close'].rolling(21).std()
X['bb_sup'] = X['sma_21'] + 2 * X['desvio']
X['bb_inf'] = X['sma_21'] - 2 * X['desvio']


X['Indicador'] = ((X['sma_21'] > X['sma_42']) & (X['sma_21'] > X['sma_200']))

X.dropna()

```

Calculando el dolar CCL, descargar el ultimo mes de cotizacion de GGAL y GGAL.BA y calcula el valor del dolar contado con liquidaciÃ³n, utilizar la columna close para calcularlo, concatenar ambos resultados y agregar dicho valor como una columna extra con el valor del calculo.
La cuenta es GGAL.BA / (GGAL * 10)

```python
ggal_ba = yf.download ('GGAL.BA', start = '2020-07-03')['Close']
ggal = yf.download ('GGAL', start = '2020-07-03')['Close']
CCL = pd.concat ([ggal_ba , ggal] , join ='outer', axis=1)
CCL.columns = ['GGAL BA','GGAL ADR']
CCL ['Dolar CCL'] = (ggal_ba/ggal*10)
CCL
```
