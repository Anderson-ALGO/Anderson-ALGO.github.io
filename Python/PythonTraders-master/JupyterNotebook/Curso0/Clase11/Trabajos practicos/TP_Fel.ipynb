```python
'''
La idea del sistema de 3 emas (4,9,18) es la sgte: 
Da señal long cuando la de 4 corta de abajo hacia arriba a la de 9 y la de 9 es mayor a la de 18.
Señal de finalizacion del trade long cuando la de 4 corta a la de 9 de arriba hacia abajo. 
De forma inversa se arman los trades short.
'''

def triple_ema_4_9_18 (ticker, start = '2018-01-01'):
    import yfinance as yf
    import pandas as pd
    
    
    
    data = yf.download(ticker, start)
    data ['ema4'] = data['Adj Close'].ewm(span = 4).mean()
    data ['ema9'] = data['Adj Close'].ewm(span = 9).mean()
    data ['ema18'] = data['Adj Close'].ewm(span = 18).mean() 
    
    data['compra'] = ((data['ema4'] > data['ema9']) & (data['ema9'] > data['ema18']))
    data['compra ant'] = ((data['ema4'] > data['ema9']) & (data['ema9'] > data['ema18'])).shift()
    data['Buy Signal'] = data['compra'] - data['compra ant']
    data['Cierre Long'] = (data['ema4'] < data['ema9'])
    
    
    data['venta'] = ((data['ema4'] < data['ema9']) & (data['ema9'] < data['ema18']))
    data['venta ant'] = ((data['ema4'] < data['ema9']) & (data['ema9'] < data['ema18'])).shift()
    data['Sell Signal'] = data['venta'] - data['venta ant']
    data['Cierre Short'] = (data['ema4'] > data['ema9'])

    data_no_0 = data[data['Buy Signal'] != 0]
    data_no_0.head()
    #data_no_0 ['Dif $'] = data_no_0['Adj Close'].shift()
      
    pd.options.display.min_rows = 100
    return data_no_0.dropna().round(2).loc[:,['Adj Close','Buy Signal','Cierre Long', 'Sell Signal', 'Cierre Short']]
    

```
