```python
def get_ticker_data(ticker):
    store = pd.HDFStore('store.h5')
    
    sp500data = store['sp500']

    data = sp500data[ticker]
    
    return pd.DataFrame(data)
```

```python
import pandas as pd
import yfinance as yf
import numpy as np

def get_trades(data, fast = 12, slow = 26, buy_cross = 0, sell_cross = 0, buy_macd = 0, sell_macd = 0, buy_rsi = 60, sell_rsi = 30):
    suavizado = 9
    ruedas=14

    data = data[['Adj Close']]
    data = pd.DataFrame(data)
    data.rename(columns={'Adj Close':'Close'}, inplace=True)
    
    sma_fast = data.Close.rolling(fast).mean()
    sma_slow = data.Close.rolling(slow).mean()
        
    data['cross'] = (sma_fast / sma_slow - 1) * 100

    ema_fast = data.Close.ewm(span=fast).mean()
    ema_slow = data.Close.ewm(span=slow).mean()

    macd = ema_fast - ema_slow
    signal = macd.ewm(span=suavizado).mean()
    data['macd'] = macd - signal

    dif = data.Close.diff()
    win = pd.DataFrame(np.where(dif > 0, dif, 0))
    loss = pd.DataFrame(np.where(dif < 0, abs(dif), 0))
    ema_win = win.ewm(alpha = 1/ruedas).mean()
    ema_loss = loss.ewm(alpha = 1/ruedas).mean()
    rs = ema_win / ema_loss   
    rs.index = data.index
    data['rsi'] = 100 - (100 / (1+rs))    
    
    data.dropna(inplace=True)

    data['signal'] = 'no-signal'
    buy = (data.cross > buy_cross) & (data.macd > buy_macd) & (data.rsi > buy_rsi)
    data.loc[buy, 'signal'] = 'buy'

    sell = (data.cross < sell_cross) | (data.macd < sell_macd) | (data.rsi < sell_rsi)
    data.loc[sell, 'signal'] = 'sell'
    
    return data.loc[data.signal != 'no-signal']

```
