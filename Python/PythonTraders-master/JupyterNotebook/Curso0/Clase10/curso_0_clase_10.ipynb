## Bandas de Bollinger

```python
import yfinance as yf

data = yf.download('GGAL')
data['sma_20'] = data['Adj Close'].rolling(20).mean()
data['volatilidad'] = data['Adj Close'].rolling(20).std()
data['boll_sup'] = data['sma_20'] + 2*data.volatilidad
data['boll_inf'] = data.sma_20 - 2*data.volatilidad
data.dropna().round(2)
```

```python
def bollinger_yahoo(ticker, sma=20, q=2):
    import yfinance as yf
    data = yf.download(ticker)
    data['sma_20'] = data['Adj Close'].rolling(sma).mean()
    desvio = data['Adj Close'].rolling(sma).std()
    data['boll_sup'] = data['sma_20'] + q*desvio
    data['boll_inf'] = data.sma_20 - q*desvio
    return data.dropna().round(2)

df = bollinger_yahoo('AMZN', sma=40)
```

## RSI

```python
import yfinance as yf
import numpy as np 

data = yf.download('YPF', end='2020-08-08')

ruedas = 14

data['dif'] = data['Adj Close'].diff()
data['win'] = np.where(data['dif'] > 0, data['dif'], 0)
data['loss'] = np.where(data['dif'] < 0, abs(data['dif']), 0)
data['ema_win'] = data.win.ewm(alpha = 1/ruedas).mean()
data['ema_loss'] = data.loss.ewm(alpha = 1/ruedas).mean()
data['rs'] = data.ema_win / data.ema_loss
data['rsi'] = 100 - (100 / (1+data.rs))
data = data.reset_index().dropna().round(2)
data
```

### Armamos funcion que vaya a buscar el excel

```python
def rsi_excel(ticker, ruedas=14):
    import pandas as pd

    data = pd.read_excel('excels_csvs/ADRs/'+ticker+'.xlsx').set_index('timestamp').sort_index()
    data['dif'] = data.adjusted_close.diff()
    data['win'] = np.where(data['dif'] > 0, data['dif'], 0)
    data['loss'] = np.where(data['dif'] < 0, abs(data['dif']), 0)
    data['ema_win'] = data.win.ewm(alpha=1/ruedas).mean()
    data['ema_loss'] = data.loss.ewm(alpha=1/ruedas).mean()
    data['rs'] = data.ema_win / data.ema_loss
    data['rsi'] = 100 - (100 / (1+data.rs))
    data = data.dropna().round(2)

    return data

rsi_excel('YPF')
```

## MACD

```python
import pandas as pd
data = pd.read_excel('excels_csvs/ADRs/YPF.xlsx').set_index('timestamp').sort_index()

slow = 26
fast = 12
suavizado = 9

data['ema_fast'] = data.adjusted_close.ewm(span=fast).mean()
data['ema_slow'] = data.adjusted_close.ewm(span=slow).mean()
data['macd'] = data.ema_fast - data.ema_slow
data['signal'] = data.macd.ewm(span=suavizado).mean()
data['histograma'] = data.macd - data.signal
data = data.dropna().round(2)
data
```

parametros:
    * ticker
    * media rapida
    * media lenta
    * suavizado
    

```python
def macd_excel(ticker, slow, fast, suavizado):
    import pandas as pd
    data = pd.read_excel('excels_csvs/ADRs/'+ticker+'.xlsx').set_index('timestamp').sort_index()
    data['ema_fast'] = data.adjusted_close.ewm(span=fast).mean()
    data['ema_slow'] = data.adjusted_close.ewm(span=slow).mean()
    data['macd'] = data.ema_fast - data.ema_slow
    data['signal'] = data.macd.ewm(span=suavizado).mean()
    data['histograma'] = data.macd - data.signal
    data = data.dropna().round(2)
    return data
```

```python
macd_excel(suavizado=9, fast=12, ticker='YPF', slow=26)
```

### MACD desde yfinance

```python
def macd_yahoo(ticker, slow, fast, suavizado):
    import yfinance as yf
    data = yf.download(ticker)
    data['ema_fast'] = data['Adj Close'].ewm(span=fast).mean()
    data['ema_slow'] = data['Adj Close'].ewm(span=slow).mean()
    data['macd'] = data.ema_fast - data.ema_slow
    data['signal'] = data.macd.ewm(span=suavizado).mean()
    data['histograma'] = data.macd - data.signal
    data = data.dropna().round(2)
    return data
```

```python
macd_yahoo('GGAL',26,12,9)
```

## Cruces

```python
def cruce_yahoo(ticker, fast=21, slow=42):

    import yfinance as yf
    data = yf.download(ticker)

    data['fast'] = data['Adj Close'].rolling(fast).mean()
    data['slow'] = data['Adj Close'].rolling(slow).mean()
    data['cruce'] = (data['fast'] / data['slow'] - 1)*100
    return data.dropna().round(2)
```

```python
cr_ypf = cruce_yahoo('YPF',50,200)
```

```python
sp500
```

### Bajamos los cierres diarios de esos tickers

```python
import yfinance as yf
import numpy as np

tickers = sp500_tickers
cierres = yf.download(tickers, start='2000-01-01')['Adj Close']
cierres = cierres.loc[:, (cierres != 0).all(axis=0)]
```

### Armamos una matriz de retornos y una de resumen

```python
retornos = np.log((cierres/cierres.shift(1)))
mat_ret = retornos.groupby(retornos.index.dayofweek).mean()*250
mat_ret = mat_ret.clip(lower=-1, upper=1)
mat_ret.index = ['Lunes','Martes','Miercoles','Jueves','Viernes']
resumen = mat_ret.transpose()
resumen.describe()
```
