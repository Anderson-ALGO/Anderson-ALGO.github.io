Les muestro varias maneras de hacer la iteraciÃ³n de los elementos de un diccionario, elijan la que quieran obviamente siempr ela mas simple y entendible es la mejor a mi criterio pero quiza por alguna razon les sirva otra en algun momento:

#### Iterando el diccionario  (Itera las claves)

```python
cartera = {"AAPL":240,"AMZN":1700,"NFLX":298,"FB":146}

cartera['AMZN']
```

```python
cartera = {"AAPL":240,"AMZN":1700,"NFLX":298,"FB":146}

for elemento in cartera:
    print('Clave:',elemento, end=', ')
    print('Valor:' , cartera[elemento])
```

#### Iterando las claves

```python
for clave in cartera.keys():
    print('Clave:',clave, end=', ')
    print('Valor:' , cartera[clave])
```

#### Iterando los valores

```python
for valor in cartera.values():
    # k = list(cartera.keys()) [list(cartera.values()).index(valor)]
    # print('Clave:',k, end=', ')
    
    print('Valor:' , valor)
```

#### Iterando pares (zip creado por nosotros)

```python
pares = list(zip(cartera.keys(), cartera.values()))

for elemento in pares:
    print('Clave:', elemento[0], end=', ')
    print('Valor:' , elemento[1])    
```

```python
for i in range(len(cartera.items())):
    print('Clave:', list(cartera.items())[i][0], end=', ')
    print('Valor:' , list(cartera.items())[i][1])    
```

## Ciclos Indefinidos Finitos
<small>
    Son ciclos en los que a priori no sabemos cuantas veces se van a ejecutar. Para esto tenemos los comandos FOR y WHILE
</small>

```python
precios = [112,113,111,114,111,119,111,101,112,115,112]
stopLoss = 110
for precio in precios:
    if(precio > stopLoss):
        print("Hold $" + str(precio))
    else:
        print("--- StopLoss at $" + str(precio) + " ---")
        break
        
```

### Ejemplo con WHILE (Nos "ahorramos" el IF, pero necesitamos un contador/indice para iterar)

```python
precios = [112,113,111,114,110,109,110,111,112,115,112]
stopLoss = 110

i=0
while precios[i] > stopLoss:
    print("Hold $" + str(precios[i]))
    i += 1
else:
    print("--- StopLoss at $" + str(precios[i]) + " ---")
    
```
