```python
#=========================#
#      Ejercicio 1        #
#=========================#

import matplotlib.pyplot as plt, yfinance as yf
df = yf .download(['YPF','CL=F'])['Adj Close']
ratio = df['CL=F']/df['YPF']
ratio_r =ratio.resample('M').mean()
ypf_r = df['YPF'].resample('M').mean()
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(ratio_r, label='ADRs x Barril')
ax.legend(loc='upper left')
ax2=ax.twinx()
ax2.plot(ypf_r, lw=0.2, c='g', label='Precio YPF')
ax2.fill_between(ypf_r.index, ypf_r, color='g',alpha=0.1)
ax2.legend(loc='upper right')
fig.suptitle('WTI/YPF', fontsize=20, y=0.95)
plt. show()
```

```python
#=========================#
#      Ejercicio 3        #
#=========================#


import requests, pandas as pd, yfinance as yf

sp500 = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
sp500_tickers = list(sp500.Symbol)
sp500_tickers = [t for t in sp500_tickers if t not in ['BRK.B' ,'BF.B']]

def getDataM(listado, start='2020-01-01', interval='1d', end=None):
    data = yf.download(listado, start=start, end=end, interval=interval, auto_adjust=True)
    return data.swaplevel(i=1, j=0, axis=1)

df_sp500 = getDataM(sp500_tickers, start='2020-01-01')

lista = []
for ticker in sp500_tickers:
    vol_mln = int((df_sp500[ticker,'Close']*df_sp500[ticker,'Volume']/1000000).mean())
    lista.append({'ticker':ticker,'vol_mln':vol_mln})

top_vol = pd.DataFrame(lista).sort_values('vol_mln', ascending=False)
top50 = list(top_vol.iloc[:50,0])
print(top50)
```

```python
#=========================#
#      Ejercicio 4        #
#=========================#

def patron(ticker):
    import yfinance as yf
    data = yf.download(ticker, auto_adjust='true')
    rango = data['High'] - data['Low']
    data['cuerpo'] = np.where(rango != 0, abs(data['Open']-data['Close']) / rango,1)
    data['gap'] = data['Open']/data['Close'].shift() - 1
    data['color_vela'] = np.where(data.Close>data.Open,'Verde','Roja')
    data['condicion'] = (data.color_vela == 'Verde') & (data.gap > 0) & (data.cuerpo > 0.4)
    data['patron'] = data.condicion & data.shift(1).condicion & data.shift(2).condicion
    return data.loc[data.patron]

patron('YPF')
```

```python
#=========================#
#      Ejercicio 5        #
#=========================#

import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

ticker = 'GGAL'
data = yf.download(ticker, auto_adjust='true')
data['gap'] = data['Open']/data['Close'].shift() - 1
data['intra'] = data['Close']/data['Open'] - 1

data['gap_vol'] = data['gap'].rolling(250).std()
data['intra_vol'] = data['intra'].rolling(250).std()

data['gap_media'] = data['gap'].rolling(250).mean()
data['intra_media'] = data['intra'].rolling(250).mean()
```

```python
#==========================#
#      Ejercicio 5-A       #
#==========================#

data['gap'].plot(kind='kde', xlim=(-0.1,0.1), label='GAPS')
data['intra'].plot(kind='kde', xlim=(-0.1,0.1), label='Intras')
plt.legend()
```

```python
#==========================#
#      Ejercicio 5-B       #
#==========================#

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(data.gap_vol, label='Desvio de los GAPS')
ax.plot(data.intra_vol, label='Desvio de los intras')
ax.legend()
```

```python
#==========================#
#      Ejercicio 5-C       #
#==========================#

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(data.gap_media, label='Promedio de los GAPS')
ax.plot(data.intra_media, label='Promedio de los intras')
ax.legend()
```
