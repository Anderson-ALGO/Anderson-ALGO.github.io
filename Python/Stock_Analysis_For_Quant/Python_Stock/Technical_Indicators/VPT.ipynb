# Volume Price Trend (VPT)

https://en.wikipedia.org/wiki/Volume%E2%80%93price_trend

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

# fix_yahoo_finance is used to fetch data 
import fix_yahoo_finance as yf
yf.pdr_override()
```

```python
# input
symbol = 'AAPL'
start = '2018-08-01'
end = '2019-01-01'

# Read data 
df = yf.download(symbol,start,end)

# View Columns
df.head()
```

```python
vpt = df['Volume'] * ((df['Adj Close'] - df['Adj Close'].shift(1)) / df['Adj Close'].shift(1))
vpt = vpt.shift(1) + vpt
vpt = vpt.replace([np.inf, -np.inf], np.nan).fillna(0)
df['VPT'] = pd.Series(vpt)
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(14,10))
ax1 = plt.subplot(2, 1, 1)
ax1.plot(df['Adj Close'])
ax1.set_title('Stock '+ symbol +' Closing Price')
ax1.set_ylabel('Price')

ax2 = plt.subplot(2, 1, 2)
ax2.plot(df['VPT'], label='Volume Price Trend')
ax2.grid()
ax2.legend(loc='best')
ax2.set_ylabel('Volume Price Trend')
ax2.set_xlabel('Date')
```

## Candlestick with Volume Price Trend

```python
from matplotlib import dates as mdates
import datetime as dt

dfc = df.copy()
dfc['VolumePositive'] = dfc['Open'] < dfc['Adj Close']
#dfc = dfc.dropna()
dfc = dfc.reset_index()
dfc['Date'] = mdates.date2num(dfc['Date'].astype(dt.date))
dfc.head()
```

```python
from mpl_finance import candlestick_ohlc

fig = plt.figure(figsize=(14,10))
ax1 = plt.subplot(2, 1, 1)
candlestick_ohlc(ax1,dfc.values, width=0.5, colorup='g', colordown='r', alpha=1.0)
ax1.xaxis_date()
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
ax1.grid(True, which='both')
ax1.minorticks_on()
ax1v = ax1.twinx()
colors = dfc.VolumePositive.map({True: 'g', False: 'r'})
ax1v.bar(dfc.Date, dfc['Volume'], color=colors, alpha=0.4)
ax1v.axes.yaxis.set_ticklabels([])
ax1v.set_ylim(0, 3*df.Volume.max())
ax1.set_title('Stock '+ symbol +' Closing Price')
ax1.set_ylabel('Price')

ax2 = plt.subplot(2, 1, 2)
ax2.plot(df['VPT'], label='Volume Price Trend')
ax2.grid()
ax2.legend(loc='best')
ax2.set_ylabel('Volume Price Trend')
ax2.set_xlabel('Date')
```
